### Cypress "Selector Playground" Feature

- This feature is used to identify the web elements using Cypress Test Runner.

- **Open the Selector Playground:** Launch the Cypress Test Runner. Within the Test Runner, locate and click the "Open Selector Playground" button, typically **positioned near the URL bar at the top**.

  

- This is the example.

  > cy.get('[data-id="41189"] > .dynamic-link')

### Running Xpath based programs in Cypress

- XPath is not directly supported in Cypress.

- To run XPath based code, first install  ```cypress-xpath``` tool

- Use the below command to install  ```cypress-xpath``` tool

> ```npm install -D cypress-xpath```

- Then use this line first in the XPath based code

> ```require('cypress-xpath')```
